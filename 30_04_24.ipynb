```python
#recursive function
def recursion(n):
  if n<=1:
    return n
  return n+recursion(n-1)
n=10
print(recursion(n))
```

```python
#fibonacci series using recursive function
def recur_fib(n):
  if n<=1:
    return(n)
  else:
    return(recur_fib(n-1)+recur_fib(n-2))
n=5
for i in range(n):
  print(recur_fib(i))
```

```python
#factorial using recursive function
def recur_fact(n):
  if n==0 or n==1:
    return 1
    recur_fact(n-1)
  else:
    return n*recur_fact(n-1)
number=5
print(recur_fact(number))
```

```python
#generating function
def gen_num(n):
  v=0
  while v<n:
    yield v
    v+=1
for v in gen_num(100):
  print(v)
```

```python
def count_case(x):
    upper = 0
    lower = 0
    for char in x:
        if char.isupper():
            upper += 1
        else:
            lower += 1
    return "upper case count", upper,"lower case count",lower

print(count_case("She sells Seashells by the Seashore"))
```
